Trajectory of charged in magnetic field if the initial velocity of the charged particle is perpendicular to the magnetic field

```python
import math as mt
import numpy as np
import matplotlib.pyplot as plt

# write the function two calculate sine and cosine arrays
def calculate_cos(angle_degrees):
    # Convert angle from degrees to radians using NumPy (handles arrays)
    angle_radians = np.radians(angle_degrees)
    # Calculate the sine of the angle using NumPy (handles arrays)
    cos_value = np.cos(angle_radians)
    return cos_value

def calculate_sin(angle_degrees):
    # Convert angle from degrees to radians
    angle_radians = np.radians(angle_degrees)
    # Calculate the sine of the angle
    sin_value = np.sin(angle_radians)
    return sin_value

# initial position
x0 = 1   # at t = 0
y0 = 1   # at t = 0
# input charge, magnetic field and mass of charged particle
q = 2 # in coulomb
B = 2 # in Tesla
m = 9.1e-2 # mass of electron
# initial velocity of electron
v = 2 # in m/s
# time array
t = np.linspace(0, 10, 100)
#print('t:', t)
omega_t = ((q * B * t)/m)
# find cos_omega_t
cos_omega_t = calculate_cos(omega_t)
#print('cos_omega_t:', cos_omega_t)
x = x0 + ((m*v)/(q*B))*(1 - cos_omega_t)
#print('x:', x)
# find sin_omega_t
sin_omega_t = calculate_sin(omega_t)
y = y0 + ((m*v)/(q*B))*(sin_omega_t)
#print('y:', y)

# plot the result
plt.plot(x, y)
plt.title('trajectory of charged particle')
plt.xlabel('x-coordinate')
plt.ylabel('y-coordinate')
plt.xlim([0.9, 1.2])  # Set x-axis limits
plt.ylim([0.9, 1.2])  # Set y-axis limits
plt.grid(True)
plt.show()
```
